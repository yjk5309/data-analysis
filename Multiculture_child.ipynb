### 검정
* 서비스를 이용할 수 있는 인구(다문화자녀)의 사용 언어 순위

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('다문화자녀.csv', engine='python', encoding='utf-8')
df
```

```python
df_sample = df.iloc[:,[0,13]]
df_sample = df_sample.drop(0)
df_sample
```

```python
df_sample['2019'] = df_sample['2019'].str.replace(',', '').astype('int64')

df_sample
```

```python
df_rank = df_sample.sort_values(by='2019', ascending=False)
df_rank
```

```python
plt.rc('font', family='Malgun Gothic')

sns.barplot(data=df_rank, x = '국적별', y = '2019')
plt.xticks(rotation=45)

plt.show()
```
